```python
epoch_acc = []
epoch_loss = []

with open('slurm-111978.out', 'r') as file:
    for line in file:
        flag = False
        
            
        if line.startswith('Accuracy'):
            
            
            try:
                acc = float(line.split('Accuracy: ')[1].split(' |')[0])
                loss = float(line.split('Loss: ')[1])
            except:
                continue
            epoch_acc.append(acc)
            epoch_loss.append(loss)
            flag = False


# makes plots nice
def make_pretty(ax, title='', x_label='', y_label='', img=False):
    ax.set_title(title, fontsize=20)
    ax.set_xlabel(x_label, fontsize=16)
    ax.set_ylabel(y_label, fontsize=16)
    ax.legend(loc='best', fontsize=16)
    if img:
        ax.axis('off')
    return ax
        

print('Accuracies per epoch:', epoch_acc)
print('Losses per epoch:', epoch_loss)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(np.arange(1, 50001), epoch_acc)

make_pretty(ax, 'Accuracy vs Epoch', 'Epoch', 'Accuracy')
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(np.arange(1, 50001), epoch_loss)
make_pretty(ax, 'Loss vs Epoch', 'Epoch', 'Loss')
```
